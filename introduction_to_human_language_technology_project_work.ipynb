### 2.2. Preprocessing
I'll be trying out SVM with TF-IDF vectorizer. Paper by Amulya, K. et al had well defined pre-processing and best results if BERT is not included. BERT uses pre-trained models and I want to try a model that only uses the given dataset. I will be following the process defined in the paper by Amulya, K. et al.

```python
# Text stemming
# Exept Lemmatizing is generally better so let do that instead
# https://www.geeksforgeeks.org/python-lemmatization-with-nltk/
en_core_web_sm = spacy.load("en_core_web_sm")

def lemmatize(ex):
  text = en_core_web_sm(ex['text'])
  lemmatized_text = " ".join([token.lemma_ for token in text])
  return{"text":lemmatized_text}

train_dset=train_dset.map(lemmatize,num_proc=4)
test_dset=test_dset.map(lemmatize,num_proc=4)
```

```python
# Text normalization: text to lower case
# Removing noisy text: HTML syntax, remove digits, remove url
# Remove punctuations

# Expanding contractions is not needed. Contractions are removed
# when stemming and stop word removal is done.

def clean_text(ex):
    text = ex["text"]
    text = text.lower() # All text to lower case
    text = re.sub(r'\d+', '', text)  # Remove digits
    text = re.sub(r'https?://\S+|www\.\S+', '', text)  # Remove URLs
    text = re.sub(r'<.*?>', '', text)  # Remove HTML tags
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    return {"text": text}

train_dset=train_dset.map(clean_text,num_proc=4)
test_dset=test_dset.map(clean_text,num_proc=4)
```

```python
pprint.pp(train_dset[0])
```

```python
train_texts = [item['text'] for item in train_dset]
test_texts = [item['text'] for item in test_dset]
train_labels = [item['label'] for item in train_dset]
test_labels = [item['label'] for item in test_dset]
```

```python
# Define stop words
# Extend the default NLTK stop words list with IMDB specific stop words
stop_words = list(set(stopwords.words('english')).union({
    'movie', 'film', 'cinema', 'director', 'actor', 'actress', 'cast',
    'plot', 'scene', 'scenes', 'character', 'characters', 'performance',
    'script', 'storyline', 'dialogue', 'screen', 'watch', 'watched',
    'watching', 'show', 'shows', 'series', 'sequel', 'prequel', 'franchise',
    'genre', 'release', 'released', 'production', 'studio', 'theaters',
    'runtime', 'minute', 'minutes', 'hour', 'hours', 'part', 'parts',
    'version', 'versions', 'hollywood', 'bollywood', 'award', 'awards',
    'oscar', 'oscars', 'sound', 'soundtrack', 'effects', 'visual', 'special',
    'special effects', 'edit', 'editing', 'directed', 'acting', 'shot', 'shots',
    'storyline', 'storylines', 'review', 'reviews', 'reviewed', 'rating', 'ratings',
    'video', 'videos', 'rent', 'store',
}))
```

```python
# Vectorize
# https://www.kaggle.com/code/lakshmi25npathi/sentiment-analysis-of-imdb-movie-reviews
tv=TfidfVectorizer(stop_words=stop_words, max_features=5000)

tv_train=tv.fit_transform(train_texts)
tv_test=tv.transform(test_texts)
```

```python
print(tv.vocabulary_)
```

## 3. Machine learning model

### 3.1. Model training

```python
# https://medium.com/@bedigunjit/simple-guide-to-text-classification-nlp-using-svm-and-naive-bayes-with-python-421db3a72d34
model = SVC(C=1.0, kernel='linear', degree=3, gamma='auto')
model.fit(tv_train,train_labels)
```

```python
predictions = model.predict(tv_test)
print("SVM Accuracy Score:", accuracy_score(predictions, test_labels))
```

### 3.2 Hyperparameter optimization



```python
# https://medium.com/all-things-ai/in-depth-parameter-tuning-for-svc-758215394769

# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100],
              'gamma': [1, 0.1, 0.01, 0.001],
              'kernel': ['rbf']}

grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3, cv=3)

# fitting the model for grid search
grid.fit(tv_train, train_labels)
```

```python
print(grid.best_params_)
print(grid.best_estimator_)

#               precision    recall  f1-score   support
#
#           0       0.88      0.86      0.87     12500
#           1       0.87      0.88      0.87     12500
#
#    accuracy                           0.87     25000
#   macro avg       0.87      0.87      0.87     25000
#weighted avg       0.87      0.87      0.87     25000
```

```python
# defining parameter range
# NOTE: different kernel
param_grid = {'C': [0.1, 1, 10, 100],
              'gamma': [1, 0.1, 0.01, 0.001],
              'kernel': ['linear']}

grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3, cv=3)

# fitting the model for grid search
grid.fit(tv_train, train_labels)
```

```python
print(grid.best_params_)
print(grid.best_estimator_)
```

### 3.3. Evaluation on test set

```python
grid_predictions = grid.predict(tv_test)
print(classification_report(test_labels, grid_predictions))
```

```python
# Let's test how the modle does in the data
# when it's not pre-processed in any way
dset = datasets.load_dataset("stanfordnlp/imdb", split=['train', 'test'])

train_dset = dset[0]
test_dset = dset[1]

train_texts = [item['text'] for item in train_dset]
test_texts = [item['text'] for item in test_dset]
train_labels = [item['label'] for item in train_dset]
test_labels = [item['label'] for item in test_dset]

tv=TfidfVectorizer(stop_words=None, max_features=5000)

tv_train=tv.fit_transform(train_texts)
tv_test=tv.transform(test_texts)

model = SVC(C=100, kernel='rbf', degree=3, gamma=0.001)
model.fit(tv_train,train_labels)

predictions = model.predict(tv_test)
print(classification_report(test_labels, predictions))
```
